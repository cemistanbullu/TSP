```python
import pandas as pd
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("greedy.csv",header=None)
df.columns=["City_no","x","y"]
df=df[["x","y"]]

geometry = [Point(xy) for xy in zip( df["y"],df["x"])]
gdf = GeoDataFrame(df, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(30, 60)), marker='o', color='red', markersize=1,alpha=0.5);
plt.plot(gdf["y"], gdf["x"],alpha=0.5)
plt.title("greedy")
plt.show()
```
